```python
#отбираем 5 лучших признаков
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest
anova = SelectKBest(f_classif, k=5)
X_train, X_test, y_train, y_test = train_test_split(df[df.columns[1:-1]], np.log(df.WORK_LIFE_BALANCE_SCORE), test_size=0.2)
X_train_anova = anova.fit_transform(X_train, y_train)
X_test_anova =  anova.fit_transform(X_test, y_test)
model = LinearRegression()
model.fit(X_train_anova, y_train)
y_pred = model.predict(X_test_anova)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
